# Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

# clustering

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans
```

```python
# by default value of cluster is 8.
model=KMeans(n_clusters=8)
```

```python
model.fit(df_new)
```

```python
model.inertia_
```

```python
model.labels_
```

```python
len(range(1,9))
```

```python
wcss=[]
for k in range(1,9):
    model=KMeans(n_clusters=k)
    model.fit(df_new)
    wcss.append(model.inertia_)
```

```python
k=np.arange(1,9)
k
wcss
```

```python
plt.figure(figsize=(5,3))
plt.scatter(k,wcss)
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.plot(k,wcss,marker="o",linestyle="--",c="r")
plt.xlabel("no of cluster")
plt.ylabel("inertia")
plt.title(" k vs wcss")
plt.show()
```
